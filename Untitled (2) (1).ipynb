# Logistic Regression  on LOAD DIGITS dataset from scratch

#### importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
import sklearn
```

#### Loading Dataset

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
Y=digits['target']
images=digits['images']
```

#### One hot Encoding

```python
Y = np_utils.to_categorical(Y,10)
```

```python
images=images/16
```

```python
plt.imshow(images[99])
Y[99]
```

#### Flattening

```python
images_new=images.reshape(images.shape[0],-1)
print(images_new.shape,Y.shape)
```

#### Train Test Split

```python
train_X,test_X,train_Y,test_Y=train_test_split(images_new,Y,test_size=0.3)
```

```python
train_X=train_X.T
test_X=test_X.T
train_Y=train_Y.T
test_Y=test_Y.T
```

#### Randomly intializing weights and bais

```python
W=np.random.randn(train_X.shape[0],train_Y.shape[0])
b=np.zeros((train_Y.shape[0],1))
```

```python
print(W.shape)
print(b.shape)
```

## LOGISTIC REGRESSION SGD with momentum

```python
v1 = 0
v2 = 0
mu = 0.99
learning_rate = 1
```

```python
for i in range(2000):
    m=train_X.shape[1]
    train_A=1/(1+np.exp(-(np.dot(W.T,train_X)+b)))
    
    # column normalization
    for j in range(train_X.shape[1]):
        train_A[:,j]=train_A[:,j]/np.sum(train_A[:,j])
    cost1=1/m*(- np.sum(train_Y* np.log(train_A+1e-6)))
    print('iteration=',i,'-----train logloss---------',cost1)    
    #cost=sklearn.metrics.log_loss(train_Y,A)
    
    test_A=1/(1+np.exp(-(np.dot(W.T,test_X)+b)))
    for j in range(test_X.shape[1]):
        test_A[:,j]=test_A[:,j]/np.sum(test_A[:,j])
    cost2=1/m*(- np.sum(test_Y* np.log(test_A+1e-6)))
    print('iteration=',i,'-----test logloss---------',cost2)          
        
    dW=(1/m)*np.dot(train_X,(train_A-train_Y).T)
    db=(1/m)*np.sum(train_A-train_Y)
    
    v1 = mu * v1 - learning_rate * dW # integrate velocity
    W += v1 # integrate position
    
    v2 = mu * v2 - learning_rate * db # integrate velocity
    b += v2 # integrate position
```

```python
#checking accuracy
count=0
for i in range(test_Y.shape[1]):
    if(np.argmax(test_Y[:,i])==np.argmax(test_A[:,i])):
        count+=1
accuracy=count/test_Y.shape[1]*100
print(accuracy)
```

## LOGISTIC REGRESSION WITH ADAM

```python
eps  = 1e-8
beta1 = 0.9
beta2 = 0.999

learning_rate = 0.001
m1 = 0 
v1 = 0

m2 = 0
v2 = 0
```

```python
W=np.random.randn(train_X.shape[0],train_Y.shape[0])
b=np.zeros((train_Y.shape[0],1))
```

```python
for i in range(2000):
    m=train_X.shape[1]
    train_A=1/(1+np.exp(-(np.dot(W.T,train_X)+b)))
    
    # column normalization
    for j in range(train_X.shape[1]):
        train_A[:,j]=train_A[:,j]/np.sum(train_A[:,j])
    cost1=1/m*(- np.sum(train_Y* np.log(train_A+1e-6)))
    print('iteration=',i,'-----train logloss---------',cost1)    
    #cost=sklearn.metrics.log_loss(train_Y,A)
    
    test_A=1/(1+np.exp(-(np.dot(W.T,test_X)+b)))
    for j in range(test_X.shape[1]):
        test_A[:,j]=test_A[:,j]/np.sum(test_A[:,j])
    cost2=1/m*(- np.sum(test_Y* np.log(test_A+1e-6)))
    print('iteration=',i,'-----test logloss---------',cost2)
    
    dW=(1/m)*np.dot(train_X,(train_A-train_Y).T)
    db=(1/m)*np.sum(train_A-train_Y)
    
    m1 = beta1*m1 + (1-beta1)*dW
    mt1 = m1 / (1-beta1**(i+1))
    v1= beta2*v1 + (1-beta2)*(dW**2)
    vt1 = v1 / (1-beta2**(i+1))
    W += - learning_rate * mt1 / (np.sqrt(vt1) + eps)
    
    m2 = beta1*m2 + (1-beta1)*db
    mt2 = m2 / (1-beta1**(i+1))
    v2= beta2*v2 + (1-beta2)*(db**2)
    vt2 = v2 / (1-beta2**(i+1))
    b += - learning_rate * mt2 / (np.sqrt(vt2) + eps)
```

```python
#checking accuracy
count=0
for i in range(test_Y.shape[1]):
    if(np.argmax(test_Y[:,i])==np.argmax(test_A[:,i])):
        count+=1
accuracy=count/test_Y.shape[1]*100
print(accuracy)
```
